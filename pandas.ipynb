```python
retail=pd.read_csv(r"C:\Users\vedap\Downloads\OnlineRetail.csv",encoding='latin1')
print(retail.head())
```

```python
retail.tail()
```

```python
retail.info()
```

```python
retail.shape
```

```python
retail.isnull().sum()
```

```python
drop_retail=retail.dropna(subset=['CustomerID'])
print(drop_retail)
```

```python
replace_retail=retail.fillna("Unknown")
print(replace_retail)
```

```python
retail['InvoiceDate'] = pd.to_datetime(retail['InvoiceDate'], errors='coerce', format='mixed')
print(retail)
```

```python
retail_dropdate = retail.dropna(subset=['InvoiceDate'])
print(retail_dropdate)
```

```python
negative_qty = retail[retail['Quantity'] < 0]
#retail = retail[retail['Quantity'] >= 0]
print(negative_qty)
```

```python
bad_prices = retail[retail['UnitPrice'] <= 0]
#retail = retail[retail['UnitPrice'] > 0]
print(bad_prices)
```

```python
duplicates = retail[retail.duplicated(subset=['InvoiceNo', 'StockCode'], keep=False)]
print(duplicates)
```

```python
retail = retail.drop_duplicates(subset=['InvoiceNo', 'StockCode'], keep='first')
print(retail)
```

```python
filter_retail=retail[retail['Quantity']>1000]
print(filter_retail)
```

```python
retail['TotalPrice']=retail['UnitPrice']*retail['Quantity']
sorted_retail=retail.sort_values(by='TotalPrice',ascending=False)
print(sorted_retail)
```

```python
product_info = pd.DataFrame({
    'StockCode': ['10001', '10002', '10003', '10004'],
    'Category': ['Stationery', 'Electronics', 'Furniture', 'Toys'],
    'Supplier': ['SupplierA', 'SupplierB', 'SupplierC', 'SupplierD']
})
retail = retail.merge(product_info, on='StockCode', how='left')
print(retail)
```

```python
country_summary = retail.groupby('Country').agg(
    TotalRevenue=('TotalPrice', 'sum'),
    MeanRevenue=('TotalPrice', 'mean'),
    OrderCount=('InvoiceNo', 'nunique')
).reset_index()
print(country_summary)
```

```python
bins = [0, 5, 20, retail['UnitPrice'].max()]
labels = ['Cheap', 'Medium', 'Expensive']
retail['PriceRange'] = pd.cut(retail['UnitPrice'], bins=bins, labels=labels, include_lowest=True)
print(retail[['UnitPrice', 'PriceRange']].head())
```

```python
price_summary = retail.groupby('PriceRange').agg(
    Purchases=('InvoiceNo', 'count'),
    AvgRevenue=('TotalPrice', 'mean')
).reset_index()
print(price_summary)
```

```python
corr_matrix = retail[['Quantity', 'UnitPrice', 'TotalPrice']].corr()
print(corr_matrix)
```

```python
retail=retail.rename(columns={'StockCode':'Product_ID','UnitPrice':'Price'})
print(retail)
```

```python
retail['Description'] = retail['Description'].str.lower()
glass_items = retail[retail['Description'].str.contains('glass', na=False)]
mug_items = retail[retail['Description'].str.contains('mug', na=False)]
glass_mug_items = retail[retail['Description'].str.contains('glass|mug', case=False, na=False)]
print(retail)
```

```python
numeric_summary = retail.describe()
print(numeric_summary)
```

```python
val_counts = retail[['Country','Category']].value_counts()
print(val_counts)
```

```python
retail['TotalPrice'] = retail['Quantity'] * retail['Price']
high_value_orders = retail.query("TotalPrice > 1000 and Country == 'Germany'")
print(high_value_orders.head())
```

```python
medium_quantity_orders = retail[retail['Quantity'].between(10, 100)]
print(medium_quantity_orders.head())
```
