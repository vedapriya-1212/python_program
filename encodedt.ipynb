```python
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder

data = ['junior', 'manager', 'senior', 'junior', 'manager', 'senior', 'junior', 'manager', 'senior', 'junior',
        'manager', 'senior', 'junior', 'manager', 'senior', 'junior', 'manager', 'senior', 'junior', 'manager']
s = pd.Series(data)
print("Original Series:")
print(s)
```

```python
label_encoder = LabelEncoder()
label = label_encoder.fit_transform(s)
print("Label Encoded Values:")
print(label)
```

```python
onehot = pd.get_dummies(s, prefix="Category").astype(int)
print("One-Hot Encoded DataFrame:")
print(onehot)
```

```python
df=pd.read_csv(r"C:\Users\vedap\Downloads\loan_approved.csv")
df.head(5)
```

```python
df.info()
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns
print(numeric_cols)
```

```python
categorical_cols=df.select_dtypes(include=['object','category']).columns
print(categorical_cols)
```

```python
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])
```

```python
df.head(14)
```

```python
df.tail(20)
```

```python
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
```

```python
df_onehot_gender = pd.get_dummies(df['Gender'], prefix="categorical").head(5).astype(int)
print(df_onehot_gender)
```

```python
df['Property_Area']=le.fit_transform(df['Property_Area'])
```

```python
df['Loan_Amount_Term']=le.fit_transform(df['Loan_Amount_Term'])
```

```python
df.to_csv(r'C:\Users\vedap\Downloads\preprocessed_loan.csv', index=False)
```
